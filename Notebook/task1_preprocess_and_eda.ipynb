```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series, series_name):
    result = adfuller(series.dropna())
    print(f"ADF Test for {series_name}:")
    print(f"  Test Statistic: {result[0]:.4f}")
    print(f"  p-value: {result[1]:.4f}")
    print(f"  Used lag: {result[2]}")
    print(f"  Number of observations: {result[3]}")
    print("  Critical Values:")
    for key, value in result[4].items():
        print(f"    {key}: {value:.4f}")
    if result[1] < 0.05:
        print(f"Result: The {series_name} series is likely stationary.")
    else:
        print(f"Result: The {series_name} series is likely non-stationary.")
    print("\n")

# Test on Closing Price
adf_test(tsla['Close'], 'Closing Price')

# Test on Daily Returns
adf_test(tsla['Daily_Return'], 'Daily Returns')
```

```python
# Fit model on training data
arima_model.fit(train)

# Forecast for the length of the test set
n_periods = len(test)
arima_forecast = arima_model.predict(n_periods=n_periods)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train.values.reshape(-1,1))
test_scaled = scaler.transform(test.values.reshape(-1,1))

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30  # e.g., use past 30 days to predict next

X_train, y_train = create_sequences(train_scaled, seq_length)
X_test, y_test = create_sequences(test_scaled, seq_length)
```
